# **Business Problem**

#### **Context**

Dalam industri e-commerce yang terus berkembang dan kompetitif, mempertahankan pelanggan menjadi lebih penting daripada sekadar menarik pelanggan baru. Tingginya tingkat *customer churn*—yakni pelanggan yang berhenti berbelanja dalam periode tertentu—dapat berdampak langsung pada pendapatan, pertumbuhan, dan keberlanjutan bisnis. Namun, tidak semua perusahaan memiliki strategi yang efektif untuk mengelola atau menekan angka churn tersebut.

Padahal, biaya untuk memperoleh pelanggan baru bisa lima kali lebih tinggi dibandingkan mempertahankan pelanggan lama dan 16 kali lebih tinggi untuk membawa pelanggan baru setara dengan pelanggan lama yang sudah ada. Pelanggan setia juga lebih mungkin melakukan pembelian berulang, merekomendasikan merek, dan memberikan kontribusi pendapatan yang lebih besar dalam jangka panjang. Oleh karena itu, penting bagi e-commerce untuk memahami penyebab churn, mengidentifikasi pelanggan berisiko, dan menerapkan strategi retensi yang proaktif.

(source: https://www.superoffice.com/blog/reduce-customer-churn/)

Salah satu pendekatan yang dapat dilakukan adalah membangun model machine learning untuk memprediksi pelanggan yang berpotensi churn. Model ini memungkinkan tim CRM dan pemasaran mengenali pelanggan berisiko lebih awal dan mengambil tindakan seperti memberikan penawaran khusus atau promosi yang dipersonalisasi. Dengan demikian, perusahaan dapat:

* Mengurangi tingkat churn
* Meningkatkan loyalitas dan kepuasan pelanggan
* Menghemat biaya akuisisi pelanggan baru
* Menyusun strategi bisnis yang lebih tepat sasaran

Data perilaku, demografi, serta riwayat transaksi dan interaksi pelanggan yang sudah tersedia menjadi modal utama dalam membangun model prediktif ini. Dengan memanfaatkannya secara tepat, perusahaan dapat mempertahankan pelanggan bernilai dan mendorong pertumbuhan bisnis secara berkelanjutan.


**Target (Label)**

* `0` : Pelanggan **tidak** churn (tetap menggunakan layanan)
* `1` : Pelanggan **churn** (berhenti menggunakan layanan)


#### **Problem Statement**

**Perusahaan e-commerce** yang terus berkembang dan bertambah banyak menghadapi tantangan dalam mempertahankan pelanggan untuk terus berbelanja. Banyak pelanggan berhenti berbelanja tanpa adanya pola yang jelas, sehingga menyulitkan perusahaan untuk melakukan tindakan pencegahan yang efektif. Tingginya tingkat customer churn menyebabkan peningkatan biaya akuisisi pelanggan baru dan penurunan pendapatan perusahaan.

Jika perusahaan tidak dapat mengidentifikasi pelanggan yang berisiko churn lebih awal, maka akan sulit untuk mempertahankan mereka dengan melakukan strategi seperti memberikan penawaran khusus atau promosi yang tepat sasaran.Akan tetapi, jika pemahaman tentang pola perilaku pelanggan yang mengarah pada churn salah dilakukan, perusahaan dapat menghabiskan banyak waktu dan sumber daya untuk kampanye promosi yang tidak efektif.


#### **Goals**
Berdasarkan permasalahan tersebut, perusahaan ingin memprediksi kemungkinan seorang pelanggan akan churn atau tetap setia menggunakan layanan e-commerce mereka. Tingkat churn yang baik untuk perusahaan e-commerce berada di angka 5% (source: https://www.woopra.com/blog/churn-rate). Berbeda jauh dengan tingkat churn pelanggan saat ini yang berada di angka sekitar 20% dan dinilai cukup tinggi dan berdampak signifikan terhadap biaya akuisisi serta penurunan loyalitas pelanggan. Permasalahan ini akan diatasi menggunakan model machine learning yang mampu memprediksi kemungkinan pelanggan akan melakukan churn berdasarkan data pelanggan. Penerapan model ini diharapkan dapat **menurunkan tingkat churn sebesar 10–15%**, mengoptimalkan strategi pemasaran, serta meningkatkan efisiensi biaya dan loyalitas pelanggan dalam jangka panjang.

#### **Metric Evaluation**
image.png

* **FP (False Positive):**

  * Model memprediksi pelanggan akan churn, padahal kenyataannya tidak churn.
  * **Konsekuensi bisnis:** Perusahaan akan mengalokasikan promosi khusus (diskon, cashback, voucher, dan kampanye retensi) kepada pelanggan ini tanpa alasan yang valid.
  * **Potensi cost:** Pengeluaran promosi yang seharusnya bisa digunakan untuk pelanggan yang benar-benar berisiko churn. Rata-rata nilai pengeluaran per pelanggan = **$8** (source: (https://www.rivo.io/blog/customer-retention-cost))


* **FN (False Negative):**

  * Model memprediksi pelanggan tidak akan churn, namun kenyataannya pelanggan tersebut benar-benar berhenti berbelanja.
  * **Konsekuensi bisnis:** Perusahaan kehilangan pelanggan bernilai tanpa mengambil tindakan pencegahan, sehingga harus mengeluarkan biaya lebih besar untuk mengakuisisi pelanggan baru.
  * **Potensi cost:** Kehilangan pendapatan pelanggan aktif, dengan rata-rata pengeluaran per pelanggan = **$128** (biaya akuisisi pelanggan baru 16x lebih banyak untuk setara dengan pelanggan lama (source: https://www.superoffice.com/blog/reduce-customer-churn/))

Karena biaya akibat kehilangan pelanggan (FN) jauh lebih besar dibandingkan biaya promosi yang salah sasaran (FP), maka metrik evaluasi yang paling cocok untuk model ini adalah **F2-score**. F2-score memberi bobot lebih besar pada kemampuan model menangkap pelanggan yang benar-benar akan churn, sekaligus tetap mempertimbangkan kesalahan prediksi lainnya.

# **Data Preparation**

### Define X and y
Memisahkan fitur (input) dan target (output) dari dataset.

### Data Splitting (Train Test Split)

Membagi data train dan data test dengan rasio 80:20

```python
X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.2,
    random_state=0,
    stratify=y
)
```

# **Data Preprocessing & Feature Engineering**

* **Imputer**:

  * **Iterative**: Digunakan untuk mengisi nilai yang hilang (missing values) pada kolom numerik **'Tenure'**, **'WarehouseToHome'**, dan **'DaySinceLastOrder'**.  Metode ini berbasis pendekatan model regresi, sehingga hasil imputasi lebih akurat dibanding metode sederhana seperti median.

* **Encoding**:

  * **OneHot**: Mengubah nilai kategorikal dari kolom **'PreferedOrderCat'** dan **'MaritalStatus'** menjadi representasi biner (0/1)

* **Scaling**:

  * **RobustScaler**: Menormalisasikan nilai semua kolom numerik, metode ini menggunakan median dan interkuartil range (IQR).


```python
impute_cols = ['Tenure', 'WarehouseToHome', 'DaySinceLastOrder']
onehot_cols = ['PreferedOrderCat', 'MaritalStatus']
scale_cols = df.select_dtypes(include='number').columns.difference(impute_cols + ['Churn'])
```

```python
transformer = ColumnTransformer([
    ('iterative', IterativeImputer(random_state=0), impute_cols),
    ('OneHot', OneHotEncoder(), onehot_cols),
    ('robust', RobustScaler(), scale_cols)
], remainder='passthrough')
```

# **Modeling and Evaluation**

### Cross Validation
Mengevaluasi performa seluruh model machine learning dengan membaginya menjadi beberapa bagian data (folds), sehingga kita bisa mengukur seberapa baik model dapat menggeneralisasi ke data baru. Model dengan nilai F2-Score terbaik akan dilakukan tuning hyperparameter

```python
# Define Model
logreg = LogisticRegression(random_state=0)
knn = KNeighborsClassifier()
tree = DecisionTreeClassifier(random_state=0)

bagging = BaggingClassifier(random_state=0, estimator=knn)
rf = RandomForestClassifier(random_state=0)
adaboost = AdaBoostClassifier(random_state=0)
gradientboost = GradientBoostingClassifier(random_state=0)
xgboost = XGBClassifier(random_state=0, use_label_encoder=False, eval_metric='logloss')
lgbm = LGBMClassifier(random_state=0, objective='binary')

voting = VotingClassifier([
    ('clf1', logreg),
    ('clf2', knn),
    ('clf3', tree)
], voting="hard")

stacking = StackingClassifier([
    ('clf1', logreg),
    ('clf2', knn),
    ('clf3', tree)
], final_estimator = logreg)
```

```python
list_model = [logreg, tree, knn, voting, stacking, bagging, rf, adaboost, gradientboost, xgboost,lgbm]

f2_scorer = make_scorer(fbeta_score, beta=2)

list_score_recall_all = []
list_score_mean_recall = []
list_score_std_recall = []

for model in list_model:
    # define pipe_model
    pipe_model = Pipeline([
        ("preprocessing", transformer),
        ("modeling", model)
    ])

    # cross validation
    cv_score = cross_validate(
        estimator=pipe_model,
        X= X_train,
        y= y_train,
        cv= 5,
        scoring= f2_scorer
    )

    # # scoring rmse
    list_score_recall_all.append(cv_score["test_score"].round(2))
    list_score_mean_recall.append(cv_score["test_score"].mean())
    list_score_std_recall.append(cv_score["test_score"].std())
```

```python
df_cv = pd.DataFrame()
df_cv["model"] = list_model
df_cv["all_score"] = list_score_recall_all
df_cv["mean_score"] = list_score_mean_recall
df_cv["std_score"] = list_score_std_recall
df_cv.sort_values(by='mean_score',ascending=False)
```

Didapatkan lima model dengan nilai mean f2-score terbaik yaitu model XGBoost, LightGBM, Decision Tree, RandomForest, dan StackingClassifier. Diambil tiga model yang akan dilakukan hyperparameter tuning, yaitu XgBoost, LGBM, dan Random Forest. Model yang dipilih merupakan model yang lebih kompleks dimandingkan model terbaik lainnya sehingga lebih cocok untuk dilakukan hyperparameter tuning karena memiliki pilihan parameter yang lebih beragam, proses ini dilakukan untuk mendapatkan model terbaik

# **Hyperparameter Tuning**

Setelah dilakukan evaluasi menggunakan 5-fold cross-validation dengan metrik F2 Score, diperoleh tiga model dengan performa rata-rata terbaik. Untuk mengoptimalkan performa ketiganya, dilakukan proses hyperparameter tuning menggunakan teknik Grid Search.

```python
from imblearn.pipeline import Pipeline as ImbPipeline
```

### **XGboost**

```python
resamplers = {
    'smote': SMOTE(random_state=0, sampling_strategy='minority'),
    'ros': RandomOverSampler(random_state=0),
    'rus': RandomUnderSampler(random_state=0),
    'nearmiss': NearMiss()
}

# Pipeline
xgb = XGBClassifier(
    random_state=0,
    use_label_encoder=False,
    eval_metric='logloss'
)

pipe_xgb = ImbPipeline([
    ("preprocessing", transformer),
    ("resampling", SMOTE()),
    ("modeling", xgb)
])

# Grid
param_grid_xgb = {
    "resampling": list(resamplers.values()),
    "modeling__n_estimators": [50, 100, 150],
    "modeling__max_depth": [3, 5, 7, 10],
    "modeling__learning_rate": [0.01, 0.05, 0.1, 0.2],
    "modeling__subsample": [0.7, 0.8, 1.0],
    "modeling__colsample_bytree": [0.7, 0.9, 1.0],
    "modeling__gamma": [0, 1, 5]
}

# Search
grid_xgb = GridSearchCV(
    estimator=pipe_xgb,
    param_grid=param_grid_xgb,
    scoring=f2_scorer,
    cv=5,
    n_jobs=-1
)
```

```python
grid_xgb.fit(X_train, y_train)
```

```python
grid_xgb.best_score_
```

```python
grid_xgb.best_params_
```

```python
pd.DataFrame(grid_xgb.cv_results_).sort_values("rank_test_score")
```

### **Random Forest**

```python
# Model
rf = RandomForestClassifier(random_state=0)

# Default resampler
ros = SMOTE(random_state=0, sampling_strategy='minority')

# Resampler
resamplers = {
    'smote': SMOTE(random_state=0, sampling_strategy='minority'),
    'ros': RandomOverSampler(random_state=0),
    'rus': RandomUnderSampler(random_state=0),
    'nearmiss': NearMiss()
}

# Pipeline
pipe_model_rf = ImbPipeline([
    ("preprocessing", transformer),
    ("resampling", ros),
    ("modeling", rf)
])

# Hyperparameter grid
hyperparam_rf = {
    "resampling": list(resamplers.values()),
    "modeling__n_estimators": [100, 200],
    "modeling__max_depth": [5, 10, 20],
    "modeling__min_samples_split": [2, 5, 10],
    "modeling__min_samples_leaf": [1, 2, 4],
    "modeling__max_features": ["sqrt", "log2"]
}

# F2 scorer
f2_scorer = make_scorer(fbeta_score, beta=2)

# Grid search
grid_rf = GridSearchCV(
    estimator=pipe_model_rf,
    param_grid=hyperparam_rf,
    scoring=f2_scorer,
    n_jobs=-1,
    cv=5
)
```

### Conclusion
Fokus utama dari analisis ini adalah mengevaluasi dampak kerugian akibat **False Negative (FN)** — yaitu saat model memprediksi pelanggan tidak akan churn, padahal kenyataannya pelanggan berhenti menggunakan layanan (churn).

* **False Negative (FN):**

  * Model memprediksi pelanggan akan tetap, padahal sebenarnya churn.
  * **Estimasi kerugian:** \$128 per pelanggan
    (berdasarkan biaya akuisisi pelanggan baru untuk menggantikan pelanggan lama)
    *Source: [SuperOffice](https://www.superoffice.com/blog/reduce-customer-churn/)*


**Tanpa Menggunakan Machine Learning**

Sebelum menggunakan model prediksi, perusahaan **tidak dapat membedakan** pelanggan yang akan churn. Akibatnya, semua pelanggan dianggap berisiko churn dan langsung diberikan promosi

|            | Predicted (0) | Predicted (1) |
| :--------: | :-----------: | :-----------: |
| Actual (0) |       0       |      546      |
| Actual (1) |       0       |      107      |

* Seluruh pelanggan diberi promosi karena tidak tahu siapa yang akan churn
* Total promosi: 653 × \$8 = \$5,224
* Karena semua pelanggan churn diberi promosi, tidak ada churn terjadi (FN = 0)
* Total kerugian = biaya promosi: \$5,224

**Menggunakan ML - Sebelum Tuning**

Dengan model XGBoost sebelum tuning, model mulai bisa mengidentifikasi sebagian pelanggan yang akan churn, tapi masih terdapat cukup banyak kesalahan FN.

|            | Predicted (0) | Predicted (1) |
| :--------: | :-----------: | :-----------: |
| Actual (0) |      532      |       14      |
| Actual (1) |       35      |       71      |

* Pelanggan yang diprediksi churn (Predicted 1) = 14 + 71 = 85 → diberi promosi → tidak churn

* FN (tidak diprediksi churn tapi sebenarnya churn) = 35 → tidak diberi promosi → churn

* FP (diprediksi churn tapi sebenarnya loyal) = 14 → diberi promosi padahal tidak perlu

Kerugian:

* FP (biaya promosi sia-sia): 14 × \$8 = \$112

* FN (pelanggan benar-benar churn): 35 × \$128 = \$4,480

* Total kerugian: \$112 + \$4,480 = \$4,592

**Menggunakan ML - Setelah Tuning**

Setelah dilakukan tuning hyperparameter, model menjadi lebih akurat dalam memprediksi pelanggan yang akan churn, sehingga jumlah FN menurun drastis.

|            | Predicted (0) | Predicted (1) |
| :--------: | :-----------: | :-----------: |
| Actual (0) |      466      |       80      |
| Actual (1) |       6       |      100      |

* Pelanggan diprediksi churn = 80 + 100 = 180 → diberi promosi → tidak churn

* FN = 6 → tidak diberi promosi → tetap churn

* FP = 80 → diberi promosi padahal loyal

Kerugian:

* FP: 80 × \$8 = \$640

* FN: 6 × \$128 = \$768

* Total kerugian: \$640 + \$768 = \$1,408

**Kesimpulan & Perbandingan**

| Kondisi                    | FN Count | Kerugian (USD) |
| -------------------------- | -------- | -------------- |
| Tanpa Machine Learning     | 107      | \$5,224       |
| Dengan ML - Sebelum Tuning | 35       | \$4,480        |
| Dengan ML - Setelah Tuning | 6        | \$1,408          |

* Pengurangan kerugian dari **tanpa ML ke sebelum tuning**: \$5,224 → \$4,480
* Pengurangan kerugian dari **sebelum tuning ke setelah tuning**: \$4,480 → \$1,408
* Pengurangan total kerugian dari **tanpa ML ke setelah tuning**:
  **\$5,224 → \$1,408**
